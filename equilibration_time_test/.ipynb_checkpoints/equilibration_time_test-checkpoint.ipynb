# Equilibration Time Test Example

In order to use `rescale_velocities` in `ReplicaExchange` is necessary to have an idea of the equilibration time of the system we are using. In order to do that here we consider the example of chignolin.

```python
import numpy as np
import openmm as mm
import openmm.app as app
from openmm import unit
import pandas as pd
from openmm.unit import kilocalorie_per_mole, angstrom
import mdtraj as md
import matplotlib.pyplot as plt
```

Here we load the pdb file associated to the system. In this chase the pdb structure is already equilibrated, so that a local energy minimization is not necessary.

```python
pdb = app.PDBFile('../chi_sys.pdb')
traj = md.load('../chi_sys.pdb')

modeller = app.modeller.Modeller(pdb.topology, pdb.positions)
```

Here differents parameters for the forcefield are indicated. Those parameters should be the same used in the replica exchange symulation. 

In this chase we consider the forcefield `amber99sbildn.xml` in an explicit solvent simulation.

```python
# forcefield's parameters
forcefield=['amber99sbildn.xml','tip3p.xml']
constraints = app.HBonds
nonbonded_cutoff = 0.9*unit.nanometer
switch_distance = 0.75*unit.nanometer
nonbonded_method = app.PME
implicit_solvent = False
solvated = True
hydrogen_mass = None

# integrator's parameters
friction = 0.1 / unit.picoseconds
timestep = 2.0 * unit.femtosecond

create_system_kwargs = dict(
    removeCMMotion=True,
    nonbondedMethod=nonbonded_method,
    nonbondedCutoff=nonbonded_cutoff,
    switchDistance=switch_distance,
    constraints=constraints,
    hydrogenMass=hydrogen_mass,
    rigidWater=True,
)
```

Here we define our system from the forcefield defined before

```python
ff = app.ForceField(*forcefield)
system = ff.createSystem(modeller.topology, **create_system_kwargs)
```

First we run a simulation at T = 280K, that is the lowest temperature we want to consider in the Replica Exchange simulation. After that we save a checkpoint of the simulation that can be loaded in a second simulation in order to test the equilibration time in a swap between 280K and 400K 

```python
temperature = 280*unit.kelvin

integrator = mm.LangevinMiddleIntegrator(temperature , friction , timestep)
integrator.setConstraintTolerance(1e-7)

simulation = mm.app.Simulation(modeller.topology, system, integrator)
simulation.context.setPositions(modeller.positions)
simulation.context.setVelocitiesToTemperature(temperature)

simulation.reporters.append(app.StateDataReporter('log0.csv', 500, step=True,
        potentialEnergy=True, temperature=True, kineticEnergy=True, totalEnergy=True)) # Save every ps

simulation.step(50000) # run for 0.05 ns 

simulation.saveCheckpoint('checkpoint.xml')
```

```python
temperature = 400*unit.kelvin

integrator = mm.LangevinMiddleIntegrator(temperature , friction , timestep)
integrator.setConstraintTolerance(1e-7)

simulation = mm.app.Simulation(modeller.topology, system, integrator)
simulation.loadCheckpoint('checkpoint.xml')
simulation.context.setVelocities(simulation.context.getState(getVelocities=True).getVelocities()*np.sqrt(400/280))
#simulation.context.setVelocitiesToTemperature(np.sqrt(temperature))

simulation.reporters.append(app.StateDataReporter('log1.csv', 500, step=True,
        potentialEnergy=True, temperature=True, kineticEnergy=True, totalEnergy=True)) # Save every ps

simulation.step(100000) # run for 0.1 ns 

simulation.saveCheckpoint('checkpoint.xml')
```

```python
temperature = 280*unit.kelvin

integrator = mm.LangevinMiddleIntegrator(temperature , friction , timestep)
integrator.setConstraintTolerance(1e-7)

simulation = mm.app.Simulation(modeller.topology, system, integrator)
simulation.loadCheckpoint('checkpoint.xml')
simulation.context.setVelocities(simulation.context.getState(getVelocities=True).getVelocities()*np.sqrt(280/400))
#simulation.context.setVelocitiesToTemperature(np.sqrt(temperature))

simulation.reporters.append(app.StateDataReporter('log2.csv', 500, step=True,
        potentialEnergy=True, temperature=True, kineticEnergy=True, totalEnergy=True)) # Save every ps

simulation.step(100000) # run for 0.1 ns 
```

Here we load the results of the two simulation as a panda's dataframe and we plot temperature and energy during the epochs

```python
dfs = list()
for f in ["log0.csv", "log1.csv", "log2.csv"]:
    data = pd.read_csv(f)
    #data['file'] = f.stem
    dfs.append(data)

df = pd.concat(dfs, ignore_index=True)
```

```python
fig, ax = plt.subplots(2,2, figsize=(15,10))

df.plot(0, 1, ax=ax[0,0], c='b')
ax[0,0].set_title('Potential Energy')
ax[0,0].axvline(50000, linestyle='dashdot', color = 'black',)
ax[0,0].axvline(70000, linestyle='dashdot', color = 'black')
ax[0,0].axvline(150000, linestyle='dashdot', color = 'black',)
ax[0,0].axvline(170000, linestyle='dashdot', color = 'black')

df.plot(0,2, ax=ax[0,1], c='r', )
ax[0,1].set_title('Kinetic Energy')
ax[0,1].axvline(50000, linestyle='dashdot', color = 'black',)
ax[0,1].axvline(70000, linestyle='dashdot', color = 'black')
ax[0,1].axvline(150000, linestyle='dashdot', color = 'black',)
ax[0,1].axvline(170000, linestyle='dashdot', color = 'black')

df.plot(0,3, ax=ax[1,0], c='g')
ax[1,0].set_title('Total Energy')
ax[1,0].axvline(50000, linestyle='dashdot', color = 'black',)
ax[1,0].axvline(70000, linestyle='dashdot', color = 'black')
ax[1,0].axvline(150000, linestyle='dashdot', color = 'black',)
ax[1,0].axvline(170000, linestyle='dashdot', color = 'black')

df.plot(0,4, ax=ax[1,1], c='y')
ax[1,1].set_title('Temperature')
ax[1,1].axvline(50000, linestyle='dashdot', color = 'black',)
ax[1,1].axvline(70000, linestyle='dashdot', color = 'black')
ax[1,1].axvline(150000, linestyle='dashdot', color = 'black',)
ax[1,1].axvline(170000, linestyle='dashdot', color = 'black')

plt.show()
```
